```python
#Read Excel file locally using read_excel function
df = pd.read_excel("input1.xlsx")
df
```

```python
df.columns #display all columns
```

```python
"""
The core of the program
Gets orginal dataframe and a string within a column name to exract the specified
column.
eg: to get all column names with 'Concept Test 1', use regex='Concept Test 1'
Returns->
       'Concept Test 1 - score',
       'Concept Test 1 - time-taken (seconds)', 'Concept Test 1 - answered',
       'Concept Test 1 - correct', 'Concept Test 1- wrong',
       'Concept Test 1 - skipped' 
"""

def extract_data(orginal_df, regex):
    new_df=orginal_df.filter(like=regex)
    new_df["Name"] = orginal_df["Name"]
    new_df["Chapter Tag"] = orginal_df["Chapter Tag"]
    new_df["id"] = orginal_df["id"]
    First_column = new_df.pop("Name")
    Sec_colmn = new_df.pop("id")
    third_colmn = new_df.pop("Chapter Tag")
    new_df.insert(0, "Name", First_column)
    new_df.insert(1, "id", Sec_colmn)
    new_df.insert(2, "Chapter Tag", third_colmn)
    new_df["Test Name"] = regex
    new_df.columns = ['Name', 'Username', 'Chapter Tag', 'score',
                   'time taken', 'answered', 'correct', 'wrong', 'skipped', 'Test_Name']
    return new_df
```

```python
df1=extract_data(df, "Concept Test 1")
df1.head()
```

```python
df2=extract_data(df, "Concept Test 2")
df2.head()
```

```python
df3 = extract_data(df, "Concept Test 3")
df3.head()
```

```python
df4 = extract_data(df, "Concept Test 4")
df4.head()
```

```python
df5 = extract_data(df, "Concept Test 5")
df5.head()
```

```python
df_final1 = [df1, df2, df3, df4, df5] #create new Concept Test combined data
df_final1=pd.concat(df_final1)
df_final1=df_final1[df_final1.loc[:] != '-'].dropna() #ignore datas with "-"
df_final1.head()
```

```python
df6 = extract_data(df, "Topic Test 1")
df7=extract_data(df, "Topic Test 2")
```

```python
df_final2 = [df6, df7] #create new Topic Test combined data
df_final2 = pd.concat(df_final2)
df_final2 = df_final2[df_final2.loc[:] != '-'].dropna() #ignore datas with "-"
df_final2.head()
```

```python
df8 = extract_data(df, "Full Chapter Test 1")
df9 = extract_data(df, "Full Chapter Test 2")
```

```python
df_final3 = [df8, df9] #create new Full Chapter Test combined data
df_final3 = pd.concat(df_final3)
df_final3 = df_final3[df_final3.loc[:] != '-'].dropna() #ignore datas with "-"
df_final3.head()
```

```python
df_final=[df_final1, df_final2, df_final3] #Join all the combined tables to create the required table
df_final=pd.concat(df_final)
```

```python
#rearrange all the columns
df_final = df_final[["Name", "Username", "Chapter Tag", "Test_Name", "answered", "correct", "score", "skipped",	"time taken", "wrong"]]
# sort based on Username and Test_Name to arrange data as shown in the Output format
df_final.sort_values(["Username", "Test_Name"])
```
